```python
import pandas as pd
import warnings
```

```python
pd.set_option('display.max_columns', None) 
```

```python
warnings.filterwarnings(action='ignore')
```

```python
df=pd.read_csv('soccerDB4.csv')
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
df['datetime']=pd.to_datetime(df['datetime'],format='%Y%m%d')
```

```python
df_2013=df[(df.datetime.dt.year > 2012) & (df.datetime.dt.year < 2014)]
```

```python
df_2014=df[(df.datetime.dt.year > 2013) & (df.datetime.dt.year < 2015)]
```

```python
df_2015=df[(df.datetime.dt.year > 2014) & (df.datetime.dt.year < 2016)]
```

```python
df_2016=df[(df.datetime.dt.year > 2015) & (df.datetime.dt.year < 2017)]
```

```python
df_2017=df[(df.datetime.dt.year > 2016) & (df.datetime.dt.year < 2018)]
```

```python
df_2018=df[(df.datetime.dt.year > 2017) & (df.datetime.dt.year < 2019)]
```

```python
df_2018=df_2018.reset_index()
```

```python
df_2018=df_2018.drop(columns='index',axis=0)
```

```python
df_2018=df_2018.rename(columns={'h_Chutes a gol':'h_chutes_a_gol','a_Chutes a gol':'a_chutes_a_gol'})
```

```python
df_2018['ht_eff_of']=1
df_2018['ht_eff_of']=df_2017['ht_eff_of'].astype(float)
```

```python
df_2018['at_eff_of']=1
df_2018['at_eff_of']=df_2017['at_eff_of'].astype(float)
```

```python
contador=0
for n in range(380):
    time=df_2018['home_team'][contador]
    grupo=df_2018.groupby(df_2018['home_team']==time).sum()
    conta=grupo.home_team_score[True] /grupo.h_chutes_a_gol[True]
    df_2018['ht_eff_of'][contador]=conta
    contador+=1
```

```python
contador=0
for n in range(371):
    time=df_2018['away_team'][contador]
    grupo=df_2018.groupby(df_2018['away_team']==time).sum()
    conta=grupo.away_team_score[True] /grupo.a_chutes_a_gol[True]
    df_2018['at_eff_of'][contador]=conta
    contador+=1
```
